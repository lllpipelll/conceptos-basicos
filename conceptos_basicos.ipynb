**Conceptos Basicos**

```python
import matplotlib.pyplot as plt # Importa la librería matplotlib para crear gráficos
import numpy as np # Importa la librería numpy para operaciones numéricas

# Generate sample data for sales and advertising
np.random.seed(0) # Establece la semilla para reproducibilidad de los datos aleatorios
advertising = 100 * np.random.rand(1000) # Genera 1000 puntos de datos aleatorios entre 0 y 100 para la publicidad
sales = 2 * advertising + 10 + 20 * np.random.randn(1000) # Genera 1000 puntos de datos para las ventas con una relación lineal y ruido aleatorio

# Create a scatter plot
plt.figure(figsize=(8, 6)) # Crea una nueva figura para la gráfica con un tamaño específico
plt.scatter(advertising, sales) # Crea la gráfica de dispersión con los datos de publicidad en el eje x y ventas en el eje y
plt.xlabel('Publicidad') # Etiqueta el eje x como 'Publicidad'
plt.ylabel('Ventas') # Etiqueta el eje y como 'Ventas'
plt.title('Gráfica de dispersión de Ventas vs Publicidad') # Establece el título de la gráfica
plt.grid(True) # Muestra una cuadrícula en la gráfica
plt.show() # Muestra la gráfica generada
```
